```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Preprocessing
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    # Removing special characters
    text = re.sub(r'\W', ' ', text)
    # Lemmatization and removing stopwords
    text = ' '.join([lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words])
    return text

df['Processed Narrative'] = df['Complaint Narrative'].apply(preprocess_text)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
from wordcloud import WordCloud 

# Let's say your DataFrame has a column 'Complaint Narrative' which contains the text
narratives = df['Processed Narrative']

# Tokenize and count word occurrences
vectorizer = CountVectorizer(stop_words='english')
word_counts = vectorizer.fit_transform(narratives)
sum_words = word_counts.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)

# Top 10 words
top_words = words_freq[:10]
top_df = pd.DataFrame(top_words, columns=['Word', 'Frequency'])

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Word', y='Frequency', data=top_df)
plt.title("Top 10 most frequent words in Complaint Narratives")
plt.show()

# WordCloud
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate_from_frequencies(dict(words_freq))

plt.figure(figsize=(20, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
#plt.title("Word Cloud for Complaint Narratives")
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_features=1000)  # Adjust max_features as needed
X = vectorizer.fit_transform(df['Processed Narrative'])
```

```python
from sklearn.cluster import KMeans

# KMeans Clustering
num_clusters = 5  # Adjust the number of clusters as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
clusters = kmeans.fit_predict(X)

df['Cluster'] = clusters
```
